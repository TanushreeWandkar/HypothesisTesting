## T Test
A t-test is a type of inferential statistic which is used to determine if there is a significant difference between the means of two groups which may be related in certain features

T-test has 2 types : 1. one sampled t-test 2. two-sampled t-test.

### One-sample T-test with Python
The test will tell us whether means of the sample and the population are different

```python
pip install scipy
```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import math
np.random.seed(6)
school_ages=stats.poisson.rvs(loc=18,mu=35,size=1500)
classA_ages=stats.poisson.rvs(loc=18,mu=30,size=60)
```

#### Hypothesis Generation
#### H0 : There is no difference in the mean of school_ages and classA_ages
#### Ha /H1 : There is difference in the mean of school_ages and classA_ages

```python
## Level of significance
alpha = 0.05
```

```python
school_ages.mean()
```

```python
_,p_value = stats.ttest_1samp(a= classA_ages, popmean= school_ages.mean())
print(p_value)
```

```python
#### level of significance alpha = 0.05
### Compare p_value and alpha to Reject or Fail to reject the Null Hypothesis
if(p_value<alpha):
    print("We Reject the Null Hypothesis H0")
else:
    print("We Fail to Reject the Null Hypothesis")
```

## Two-sample T-test With Python
The Independent Samples t Test or 2-sample t-test compares the means of two independent groups in order to determine whether there is statistical evidence that the associated population means are significantly different. The Independent Samples t Test is a parametric test. This test is also known as: Independent t Test

#### Hypothesis Generation
#### H0 : There is no difference in the mean of classA_ages and classB_ages
#### Ha /H1 : There is difference in the mean of classA_ages and classB_ages

```python
np.random.seed(12)
classB_ages = stats.poisson.rvs(loc=18,mu=33,size=30)
print(classB_ages.mean())
```

```python
type(classB_ages)
```

```python
print(np.median(classB_ages))
```

```python
_, p_value_b = stats.ttest_ind(a= classA_ages,b=classB_ages)
```

```python
print(p_value_b)
```

```python
#### level of significance alpha = 0.05
### Compare p_value and alpha to Reject or Fail to reject the Null Hypothesis
if(p_value_b <alpha):
    print("We Reject the Null Hypothesis H0")
else:
    print("We Fail to Reject the Null Hypothesis")
```

## Chi Square Test
The Chi-square test is one of the statistical tests we can use to decide whether there is a correlation between the categorical variables by analysing the relationship between the observed and expected values.


Chi-squared test is a hypothesis test that is used to determine whether there is a significant association between two categorical variables in the data. The test involves two hypotheses (H0 & H1):

H0 : The two categorical variables have no relationship (independent)
H1 : There is a relationship (dependent) between two categorical variables

## We can simply use this test for feature selection to see if the two categorical variables have any relationship or correlation between them.

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(10)
## create a random set of data
bottles_type = np.random.choice(a=['paper','cans','glass','plastic','other'],size=1500)
month = np.random.choice(a=['Jan','Feb','Mar'],size=1500)
```

```python
print(bottles_type)
```

```python
print(month)
```

```python
type(bottles_type)
```

```python
##create a dataframe from this np array using a dictionary
recycle_df = pd.DataFrame({'bottle_type': bottles_type, 'month': month})
print(type(recycle_df))
print(recycle_df)
```

```python
crosstab_recycle = pd.crosstab(recycle_df.bottle_type, recycle_df.month,margins=True)
```

```python
crosstab_recycle.index = ['cans','glass','other','paper','plastic','col_total']
crosstab_recycle.columns = ['Jan','Feb','Mar','row_total']
```

```python
print(crosstab_recycle)
```

```python
observed = crosstab_recycle.iloc[0:5,0:3]
```

```python
crosstab_recycle.iloc[0:5,0:3]
```

## Implementing the Chi-square test
Let’s implement the Chi-square test and check the independence of columns. So in the formula discussed above, we need two things: observed values and expected values. So we need to understand how to calculate these values. The observed value is nothing but the original data.

How to calculate expected values?

Multiply the row total to the column total and divide by the total number of observations for a cell to get the expected count. So, to calculate the expected counts in python we need to use the np.outer() function on row total and column total. At last, dividing the output of the np.outer() function by observations.

```python
expected =  np.outer(crosstab_recycle["row_total"][0:5],
                     crosstab_recycle.loc["col_total"][0:3]) / 1500
 
expected = pd.DataFrame(expected)
 
expected.columns = ["Jan","Feb","Mar"]
expected.index = ['cans','glass','other','paper','plastic']
 
expected
```

```python
pip install scipy
```

### Calculate the chisqaure statistic

```python
chi_squared_stat = (((observed-expected)**2)/expected).sum().sum()
print(chi_squared_stat)
```

Note: We call .sum() twice, once to get the column sums and a second time to add the column sums together, returning the sum of the entire 2D table.

```python
from scipy.stats import chi2
from scipy.stats import chi2_contingency
```

```python
critical_value= chi2.ppf(q = 0.95,  df = 8) # Find the critical value for 95% confidence*  # df= degree of freedom
print("Critical value:",critical_value)
p_value = 1 - chi2.cdf(x=chi_squared_stat,df=8)  # Find the p-value
print("P value:",p_value)
```

Note: The degrees of freedom for a test of independence equals the product of the number of categories in each variable minus 1. In this case we have a 5×3 table so df = 4×2 = 8.

```python
#### level of significance alpha = 0.05
### Compare p_value and alpha to Reject or Fail to reject the Null Hypothesis
if(p_value <0.05):
    print("We Reject the Null Hypothesis H0")
else:
    print("We Fail to Reject the Null Hypothesis")
```
